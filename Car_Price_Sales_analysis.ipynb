## Correlation Heat Map

* Heatmap of the correlation between the numeric variables. Does it make sense year and condition and odomoter are negatively correlated?

```python
numeric_columns = ['year', 'condition', 'odometer', 'mmr', 'sellingprice'] 

# Create the heatmap using only the specified columns

sns.heatmap(
    cars[numeric_columns].corr(),
    cmap="RdYlGn",
    annot=True

)
```

```python
brand_list = ["Ford" , "BMW" , "Toyota" , "Chevrolet" ]

sns.lmplot(
    x = "odometer",
    y = "sellingprice",
    data=cars.query("make in @brand_list and sellingprice <150000 and odometer < 200000").sample (1000),
    hue="make",
    scatter_kws = ({"alpha" : .3 })
)
```

```python
import scipy.stats as stats

cars_regress = cars.dropna()

stats.linregress(x = cars_regress["odometer"] , y = cars_regress ["sellingprice"])
```

```python
for brand in brand_list:
    data = cars_regress.loc[cars_regress  ["make"] == brand]
    print(f"{brand}: {stats.linregress(data['odometer'] , data['sellingprice'])[0:2]}")
```

# Categorical Variable Relationships

* Bin the `condition` variable into increments of .5. For example, the first bin  is '1-1.5', the second '1.5-2', and so on up until the maximum condition value of 5. Using pd.cut() to create the bins.

* Barplot of the average price by condition bin.

```python
fig, ax = plt.subplots()

bins = np.linspace(1,5,9)
labels = ["1-1.5", "1.5-2", "2-2.5", "2.5-3", "3-3.5", "3.5-4", "4-4.5", "4.5-5"]

cars["condition_bins"] = pd.cut(cars["condition"] , bins = bins , labels = labels)

sns.barplot(
    x = "condition_bins",
    y = "sellingprice",
    data = cars,
    palette="Set2"
)

ax.set_title ("Average Price by condition")

sns.despine()
```

# Price by make and condition

* Creating a pivot table with the top 10 most common car brands `make` column as rows, and condition bins as columns. The values in the table are the average selling price for each make/condition combination. Consider using value_counts to get the top 10 most common brands.

* After the creation of the table, building a heatmap from the data. Which brands and quality levels have the highest selling prices? Does this match our expecations?

```python
top10 = cars["make"].value_counts().iloc[:10].index

avgs =  (cars
         .loc[cars["make"].isin (top10)]
         .pivot_table(index = "make",
                     columns = "condition_bins",
                     values = {"sellingprice" : "mean"}
                     )
         .droplevel(0 ,axis = 1)
        )

avgs.head()
```

```python
fig, ax = plt.subplots (figsize = (10,10))

plt.title("Avg. Selling Price by Make and Quality", fontsize=16)
sns.heatmap(
        avgs,
        annot=True,
        cmap="RdYlGn",
        fmt="g"
)
```

## Are there any "Deals"?

Now, with the help of above charts we're getting closer to honing in on the trucks to purchase.

* Creating a column `diff_to_mmr` which is the difference between `sellingprice` and `mmr`.
* A barplot looking at mean `diff_to_mmr` by `color`.
* Which color sells for more than the recommended price?

```python
sns.set_style("darkgrid")

f150s = f150s.assign(diff_to_mmr = f150s["sellingprice"] - f150s["mmr"])

f150s_by_color = (
    f150s
    .groupby("color" , as_index = False)
    .agg({"diff_to_mmr" : "mean"})
)

fig, ax = plt.subplots()

sns.barplot(
    x = "color",
    y = "diff_to_mmr",
    data = f150s_by_color,
    palette="Set1"
    
)

plt.xticks(rotation = 90)

plt.show()
```

* Further filter the data to the trims below. Excludes the color 'orange' and look only at trucks from years between 2009 and 2014. 

```python
trim_list = ['XL', 'XLT', 'Platinum', 'SVT Raptor']

f150s_reduced = f150s.query("color != 'orange' and trim in @trim_list")

f150s_pivot = (
    f150s_reduced
    .loc[f150s_reduced["year"].between(2009, 2014)]
    .pivot_table(
        index="trim",
        columns="condition_bins",
        values="sellingprice",
        aggfunc="mean"
    )
)
    
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(f150s_pivot, annot=True, fmt="g")
```

## Best state to buy trucks in?

* Ok, we've decided on the XLT model, it has more features than XL but is still quite affordable. Also here filtering out any trucks with a quality of less than 3.5. Your DataFrame should only have trucks with a quality of 3.5 or greater.
* Building a subplot with a bar chart of `state` by `diff_to_mmr` and `state` by `count` to find which states sell XLT models below mmr and have ample quantity.

```python
f150s_XLT = (f150s_reduced
             .query("trim == 'XLT' and condition >=3.5")
             .groupby(['state'] , as_index = False)
             .agg(mean_diff = ("diff_to_mmr" , "mean"),
                  auto_count = ("diff_to_mmr" , "count")
                 )
            )

fig,ax = plt.subplots(2 , figsize = (8,6))

sns. barplot(
    x = "state",
    y = "mean_diff",
    data = f150s_XLT,
    palette="Set2",
    ax =ax[0]
)


sns.barplot(
    x = "state",
    y = "auto_count",
    data = f150s_XLT,
    palette="Set2",
    ax = ax[1]
)
```

* Ok, looks like Utah, `state=='ut'`, (our client has an office nearby) has a good amount of quantity and a low average price. By looking at the average diff_to_mmr and count of cars by saledate we will get the table.

```python
f150s.query("trim == 'XLT' and state == 'ut' and condition >=3.5").groupby("saledate").agg({"diff_to_mmr" : ["mean" ,"count"]})
```
